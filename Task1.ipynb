```python
import cv2
import numpy as np
import zipfile
import os
from collections import deque
from google.colab import files # I am doing this using Google Colab

MOD = 1000000007
```

```python
# Upload zip file
uploaded = files.upload()

zip_name = list(uploaded.keys())[0]
```

```python
# Extract
extract_dir = "mazes"
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_name, 'r') as z:
    z.extractall(extract_dir)
```

```python
# Collect image paths
image_paths = []

for root, dirs, files_ in os.walk(extract_dir):
    for f in files_:
        if f.lower().endswith((".png", ".jpg", ".jpeg", ".bmp")):
            image_paths.append(os.path.join(root, f))

image_paths.sort()

print("Total images found:", len(image_paths))
```

```python
# Convert the image into a proper maze
def image_to_maze(img):
    """
    Convert 800x800 image to 40x40 binary maze.
    1 = open
    0 = blocked
    """

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    maze = np.zeros((40, 40), dtype=np.int32)

    CELL = 20

    for r in range(40):
        for c in range(40):
            cell = gray[
                r*CELL:(r+1)*CELL,
                c*CELL:(c+1)*CELL
            ]

            mean_val = np.mean(cell)

            if mean_val > 127:
                maze[r][c] = 1
            else:
                maze[r][c] = 0

    return maze
```

```python
# Performing BFS on the obtained maze
def shortest_path_length(maze):
    """
    BFS shortest path.
    Returns path length including start and end cells.
    Returns None if unsolvable.
    """

    n = 40
    m = 40

    if maze[0][0] == 0:
        return None

    if maze[n-1][m-1] == 0:
        return None

    q = deque()
    q.append((0, 0, 1))

    visited = np.zeros((n, m), dtype=bool)
    visited[0][0] = True

    directions = [
        (-1, 0),
        (1, 0),
        (0, -1),
        (0, 1)
    ]

    while q:
        r, c, dist = q.popleft()

        if r == n-1 and c == m-1:
            return dist

        for dr, dc in directions:
            nr = r + dr
            nc = c + dc

            if (
                0 <= nr < n
                and 0 <= nc < m
                and not visited[nr][nc]
                and maze[nr][nc] == 1
            ):
                visited[nr][nc] = True
                q.append((nr, nc, dist + 1))

    return None


answer = 1
solvable = 0
unsolvable = 0

for idx, path in enumerate(image_paths, start=1):

    img = cv2.imread(path)

    if img is None:
        continue

    maze = image_to_maze(img)

    dist = shortest_path_length(maze)

    if dist is None:
        unsolvable += 1
        continue

    solvable += 1
    answer = (answer * dist) % MOD

    print(
        f"Maze {idx:03d}: shortest path = {dist}"
    )

print("\n====================")
print("Solvable   :", solvable)
print("Unsolvable :", unsolvable)
print("Final Answer:", answer)
print("====================")
```
